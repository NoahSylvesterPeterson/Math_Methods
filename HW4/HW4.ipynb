```python
import numpy as np

# parameters
a = 1.0
b = 2/np.pi

# number of grid points (use even N for simplicity)
N = 4096
L = 2 * np.pi
x = np.linspace(-np.pi, np.pi, N, endpoint=False)

# function
f = np.where(np.abs(x) >= 1/b, 0, 1) * a * (1 - b*np.abs(x))

# FFT (note: this corresponds to complex exponential coefficients)
F = np.fft.fft(f) / N   # normalize so coefficients match integral form

# --------------------------------------------
# Convert FFT coefficients  real Fourier series
# --------------------------------------------
a0 = 2 * F[0].real          # a0 term

a_coeff = np.zeros(10)
b_coeff = np.zeros(10)

for n in range(1, 11):
    # FFT index n corresponds to e^{inx}
    a_coeff[n-1] = 2 * F[n].real
    b_coeff[n-1] = -2 * F[n].imag   # negative sign due to FFT convention

# print results
print( f"{a0:.6f}")
for n in range(1, 10+1):
    print(f"{a_coeff[n-1]:.6f}")

y = a0*np.cos((0)*np.pi/(l)*x)/2
y1 = a0*np.cos((0)*np.pi/(l)*x)/2
for i in range(11):
    #y = y+ np.abs(a_n(i+1))*np.cos((i+1)*np.pi/(l)*x)
    y1 = y1+ a_n(i+1)*np.cos((i+1)*np.pi/(l)*x)
    y = y+ a_n(i+1)*np.cos((i+1)*np.pi/(l)*x)
plt.plot(x,y)
plt.plot(x,y1)
```
