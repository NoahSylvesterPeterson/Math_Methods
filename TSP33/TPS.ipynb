```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
fig, ax = plt.subplots()
xdata, ydata = [], []
count =0 

l=1
nn = 2047
numterms = 30
x = np.linspace(-l,l,nn)
def f(x):
    return(x)
def b_n(n):
    return((-1.0)**(n+1)*2.0*l/n/np.pi)
def a_n(n):
    return(0)

ln, = plt.plot(x, f(x)) # 'ro' for red circles
Ft, = plt.plot(x, a_n(1)*np.cos(1*np.pi/l*x)+b_n(1)*np.sin(1*np.pi/l*x)) 


def init():
    ax.set_ylim(-1.5, 1.5)
    return ln,Ft

def update(frame):
    ydata= Ft.get_ydata() + a_n(frame+1)*np.cos((frame+1)*np.pi/l*x)+ b_n(frame+1)*np.sin((frame+1)*np.pi/l*x)
    Ft.set_data(x, ydata)
    return ln,Ft

ani = FuncAnimation(fig, update, frames=np.linspace(1,numterms+1,numterms+1),
                    init_func=init, blit=True)
#plt.show()
# To save as a GIF:
ani.save('anim.gif', writer='pillow', fps=int(numterms/2))
```

```python
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'ro') # 'ro' for red circles

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True)
plt.show()
# To save as a GIF:
ani.save('OLD_anim.gif', writer='pillow', fps=30)
```
