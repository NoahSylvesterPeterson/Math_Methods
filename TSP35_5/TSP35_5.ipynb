```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.ion()
plt.close()
nn = 256
a = 5
b = 1
phi_a = 0
phi_b = 0
scale_a = 1
scale_b = 10
time = 2*np.pi*np.arange(0,nn)/nn
signal = a*np.cos(scale_a*time+phi_a)+b*np.sin(scale_b*time+phi_b)
plt.plot(time,signal)
plt.show()
plt.clf()
amp = np.fft.fft(signal)
#plt.plot(amp)
plt.show()
plt.clf()

pow = amp*np.conjugate(amp)

plt.plot(np.log10(pow))
#plt.pause(1.0)
plt.show()
```
