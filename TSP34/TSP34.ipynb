```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1024
time1 = np.linspace(0,2*np.pi,n)
time2 = 2*np.pi*np.arange(0,n)/n

s1 = np.sin(time1)
s2 = np.sin(time2)
c1 = np.cos(time1)
c2 = np.cos(time2)

r1 = s1*c1
r2 = s2*c2
print(np.sum(s1))
print(np.sum(s2))
print(np.sum(c1))
print(np.sum(c2))
print(np.sum(r1))
print(np.sum(r2))



r3 = s2*s2
print(np.sum(r3))
```
